# Lab 07: Code Execution Tool (Gemini 3.1)

Gemini 3.1 features a built-in Python interpreter. This tool allows the model to write and execute code to solve complex reasoning, math, and data processing tasks with 100% accuracy.

## Objectives
1. Enable the **Code Execution** tool.
2. Solve complex **mathematical and algorithmic** problems.
3. Perform **Data Analysis** on raw datasets.
4. **Inspect** the Python code generated by the model.

```python
import os

from dotenv import load_dotenv
from google import genai
from google.genai import types
from IPython.display import Markdown, display

load_dotenv()
client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))

print("Client initialized.")
```

## 1. Algorithmic Problem Solving

When you enable `code_execution`, Gemini can verify its reasoning by running code. Let's ask for a complex calculation that models usually struggle with if they don't use code.

```python
config = types.GenerateContentConfig(
    tools=[types.Tool(code_execution=types.ToolCodeExecution())]
)

prompt = "Calculate the sum of all prime numbers between 1 and 1000."

response = client.models.generate_content(
    model="gemini-3.1-flash-lite-preview",
    config=config,
    contents=prompt
)

display(Markdown(response.text))
```

## 2. Data Analysis on Raw Text

We can provide a raw dataset and ask Gemini to process it using standard libraries like `numpy` or `math` (which are available in the sandbox).

```python
raw_data = """
User_ID,Purchase_Amount,Date
U1,150.50,2026-01-01
U2,200.00,2026-01-02
U1,50.25,2026-01-03
U3,450.00,2026-01-04
U2,120.00,2026-01-05
U4,30.00,2026-01-06
"""

prompt = (
    f"Analyze this raw data: {raw_data}. \n\nTask: Calculate the total revenue "
    "and find the User_ID with the highest average purchase."
)

response = client.models.generate_content(
    model="gemini-3.1-flash-lite-preview",
    config=config,
    contents=prompt
)

display(Markdown(response.text))
```

## 3. Inspecting the Generated Code

It's useful to see exactly *how* Gemini solved the problem. We can inspect the response parts to find the `executable_code` and the `execution_result`.

```python
print("Execution Trace Details:\n")
for part in response.candidates[0].content.parts:
    if part.executable_code:
        print("--- GENERATED CODE ---")
        print(part.executable_code.code)
    if part.code_execution_result:
        print("\n--- EXECUTION OUTPUT ---")
        print(part.code_execution_result.output)
```
